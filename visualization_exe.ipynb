### Step 1. Import the necessary libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/07_Visualization/Titanic_Desaster/train.csv)

### Step 3. Assign it to a variable titanic 

```python
titanic = pd.read_csv('https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/07_Visualization/Titanic_Desaster/train.csv')
```

```python
titanic
```

### Step 4. Set PassengerId as the index 

```python
titanic.set_index(['PassengerId'])
```

### Step 5. Create a pie chart presenting the male/female proportion

```python
males = (titanic['Sex'] == 'male').sum()
females = (titanic['Sex']=='female').sum()
plt.pie([males,females],labels=['Male','Female'],colors=["salmon", "lightblue"],autopct='%1.1f%%',shadow = True,explode = (0.1 , 0.1),startangle=90,rotatelabels=True,frame=False)
plt.title('Gender Ratio')
plt.axis('equal')
plt.tight_layout()
plt.show
```

```python
titanic.head()
```

### Step 6. Create a scatterplot with the Fare payed and the Age, differ the plot color by gender

```python
plt.scatter(titanic.Fare[titanic.Sex=='male'],titanic.Age[titanic.Sex=='male'],c=["salmon"],plotnonfinite=False)
plt.scatter(titanic.Fare[titanic.Sex=='female'],titanic.Age[titanic.Sex=='female'],c=["lightblue"],plotnonfinite=False)
plt.tight_layout()
plt.title('price paid')
plt.xlabel('Fare')
plt.ylabel('Age')
plt.ylim((-5,85))
plt.xlim((-5,85))
plt.legend(('Male','Female'))
plt.show()
```

### Step 8. Create a histogram with the Fare payed

```python
plt.hist(titanic.Fare,bins=np.arange(0,305,2.5),color=['salmon'])
plt.title('Fare per population')
plt.xlabel('Fare')
plt.ylabel('No.of Occurrence')
plt.show()
```
